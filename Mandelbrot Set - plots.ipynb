```python
from __future__ import division
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import ipywidgets as widgets
from ipywidgets import interact
from math import log
%matplotlib inline
```

```python
a = 2
F = lambda x:x**2-a
f = lambda x:2*x
lin = lambda x0,x:F(x0)+f(x0)*(x-x0) 
```

```python
@interact(punto =widgets.FloatSlider(
    value=5,
    min=0,
    max=7,
    step=0.1,
    description='$X_0$:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f'))
def new1(punto):
    _ = plt.figure(figsize=(10,8))
    _ = plt.xlim((0,8))
    _ = plt.ylim((-3,40))
    _ = plt.plot(np.linspace(0,8,100),map(F,np.linspace(0,8,100)),label = '$f(x)$')
    _ = plt.plot([0,8],[0,0])
    _ = plt.scatter([punto],[F(punto)],color = 'red',label = "$f'(x_0)$")
    _ = plt.plot(np.linspace(0,8,100),map(lambda y:lin(punto,y),np.linspace(0,8,100)),
                 label = "$f(x_0)+f'(x_0)(x-x_0)$")
    _ = plt.legend(loc=2)
    
```

```python
from math import cos
f = lambda x:x**2+5*x-2+30*cos(x*7)
g = lambda x:x**2+4*x+55
```

```python
x = np.linspace(-2.5,2.5,200)
y1 = map(f,x)
y2 = map(g,x)
plt.figure(figsize=(10,8))
plt.plot(x,y1,label = '$f(x)$')
plt.plot(x,y2,label = '$g(x)$')
plt.xlim((-2.5,2.5))
plt.ylim((-50,100))

plt.scatter([0,5],[0,50],color='#642523')
plt.fill_between(x=x,y1=y1,y2=y2,color='grey',alpha=0.5)
plt.title('Grafica de prueba')
plt.legend()
```

```python
def Mandelbrot(a):
    z = 0
    for i in range(30):
        if abs(z)>2:
            break
        else:
            z = z**2+a
    if abs(z)>2:
        return(abs(z))
```

```python
x = np.arange(-2, 0.5, 0.001)
y = np.arange(-1.1, 1.1, 0.001)
xx, yy = np.meshgrid(x, y, sparse=True)
z = np.array([[Mandelbrot(complex(s,t)) for s in xx[0]] for t in yy[:,0]])
```

```python
plt.figure(figsize=(15,11))
h = plt.contourf(x,y,z)
plt.show()
```
