```python
from __future__ import division
import matplotlib.pylab as plt
import numpy as np 
%matplotlib inline
```

```python
def B3(x):
    if abs(x)<1:
        return( 1/6*((2-abs(x))**3 - 4*(1-abs(x))**3))
    elif abs(x)<2:
        return( 1/6*(2-abs(x))**3)
    else:
        return(0)
```

```python
plt.plot(np.linspace(-3,3,100),[B3(z) for z in np.linspace(-3,3,100)])
```

```python
def BSplines(x,y):
    n = len(x)
    h = max([abs(x[i]-x[i+1]) for i in range(n-1)])
    A = np.matrix([[B3((x[j] -x[i])/h) for j in range(n)] 
               for i in range(n)])
    alpha = np.linalg.inv(A)*np.matrix(y).T
    spline = lambda z: sum(alpha[i,0]*B3((z-x[i])/h) for i in range(n))
    plt.figure(figsize=(12,8))
    plt.plot(np.linspace(x[0],x[-1],1000),
             [spline(z) for z in np.linspace(x[0],x[-1],1000)])
    plt.scatter(x,y,color='red')
```
